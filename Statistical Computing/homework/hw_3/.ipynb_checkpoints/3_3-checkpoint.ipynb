```python
import numpy as np
```

```python
def init(prob, pi):
    pi = np.random.random(4)
    pi /= np.sum(pi)
    
    for i in range(0, 100):
        prob[i][1][1] = 1.0
        prob[i][2][1] = 1.0
    for i in range(100, 150):
        prob[i][1][1] = 1.0
        prob[i][2][2] = 1.0
    for i in range(150, 225):
        prob[i][1][2] = 1.0
        prob[i][2][1] = 1.0
    for i in range(225, 300):
        prob[i][1][2] = 1.0
        prob[i][2][2] = 1.0
    for i in range(300, 300 + 28):
        prob[i][1][1] = 1.0
    for i in range(300 + 28, 300 + 28 + 60):
        prob[i][1][2] = 1.0
    for i in range(300 + 28 + 60, 300 + 28 + 60 + 30):
        prob[i][2][1] = 1.0
    for i in range(300 + 28 + 60 + 30, 300 + 28 + 60 + 30 + 60):
        prob[i][2][2] = 1.0
    return prob, pi
```

```python
def e_step(prob, pi):
    for i in range(300, 300 + 28):
        prob[i][2][1] = pi[0] / (pi[0] + pi[1])
        prob[i][2][2] = pi[1] / (pi[0] + pi[1])
    for i in range(300 + 28, 300 + 28 + 60):
        prob[i][2][1] = pi[2] / (pi[2] + pi[3])
        prob[i][2][2] = pi[3] / (pi[2] + pi[3])
    for i in range(300 + 28 + 60, 300 + 28 + 60 + 30):
        prob[i][1][1] = pi[0] / (pi[0] + pi[2])
        prob[i][1][2] = pi[2] / (pi[0] + pi[2])
    for i in range(300 + 28 + 60 + 30, 300 + 28 + 60 + 30 + 60):
        prob[i][1][1] = pi[1] / (pi[1] + pi[3])
        prob[i][1][2] = pi[3] / (pi[1] + pi[3])
    return prob
```

```python
def m_step(prob):
    pi_new = np.zeros(shape=4)
    for i in range(478):
        for j in range(2):
            for k in range(2):
                pi_new[j * 2 + k] += prob[i][j + 1][k + 1]
    pi_new /= np.sum(pi_new)
    return pi_new
```

```python
def em_step(prob, pi):
    prob = e_step(prob, pi)
    pi = m_step(prob)
    return prob, pi
```

```python
prob = np.zeros(shape=[478, 3, 3])
pi = np.zeros(shape=4)
prob, pi = init(prob, pi)
for i in range(500):
    prob, pi = em_step(prob, pi)
    print(pi)
```
